```python
books = [
    {'name' : 'The Jewel', 'read' : True},
    {'name' : 'Cinder', 'read' : False},
    {'name' : 'The Train to Impossible Places', 'read' : False}
]

def list_books():
    for index, book in enumerate(books):
        print(str.format('{} : {} (Completed: {})', index, book['name'], book['read']))
        
def add_book():
    book_text = input('Please add a book:')
    new_book = {'name': book_text, 'read': False}
    books.append(new_book)
    
def remove_book():
    list_books()
    remove_book = int(input('Please remove a book'))
    books.remove(books[remove_book])
    
def read_book():
    list_books()
    read_book = int(input("Read a book:"))
    books[read_book]['read'] = True
        
menu_text = """
====================
1. List the books
2. Add a book
3. Remove a book
4. Mark a book read
5. Quit

What would you like to do? """

program_is_running = True

while program_is_running:
    decision = input(menu_text)
    if decision == '1':
        list_books()
    
    if decision == '2':
        add_book()
        
    if decision == '3':
        remove_book()
        
    if decision == '4':
        read_book()
            
    elif decision == '5':
        program_is_running = False
    else:
        print('please choose a valid option')
```

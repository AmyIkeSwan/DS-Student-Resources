### DSO110 Final Project Example for Graphs

#### Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
import matplotlib.mlab as mlab
```

#### Import Data

```python
HeartFail = pd.read_csv('../GitHub/Final_Project/Datasets/heart_failure_clinical_records_dataset.csv')
```

```python
HeartFail.head()
```

#### Data Source

```python
### Data from UCI: https://archive.ics.uci.edu/ml/datasets/Heart+failure+clinical+records
```

#### Correlation Matrix

```python
HeartFail.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
### This code creates a correlation matrix, you can copy and paste this to create one for your dataset! Just change the name
### 'HeartFail' to the name of your dataset!
```

#### Heat Map

```python
plt.figure(figsize=(16,14))
sns.heatmap(HeartFail.corr(), cmap ="YlGnBu",linewidths = .1, annot = True)
### This is a heat map serving the same purpose as the correlation matrix, looking at how correlated variables are. This code uses
### the seaborn package to create a heat map. Copy and paste it to create a heat map for your dataset and change 'HeartFail'.corr
### to the name of your dataset.corr() in the second line.
```

```python
### You can see that the darker blue color shows a higher correlation, so here, gender and smoking are highly correlated at 0.45.
### And you can go along the bottom line showing DEATH_EVENT to see which variables were highly correlated.
```

#### Cluster Map

```python
plt.figure(figsize=(26,18))
sns.clustermap(HeartFail.corr(), cmap ="YlGnBu",linewidths = .5, annot = True)
### This code shows a cluser map which groups together correlated variables. You can copy and paste this code to create your
### own cluster map and just change the dataset name 'HeartFail' to match the name of your dataset:)
```

```python
### Do you see how the top left corner shows a cluster of darker squares? The cluster map "clusters" the varibles that are more 
### correlated, here you see age, serum creatinine, and death event more strongly correlated, as well as the bottom right corner
### showing ejection fraction and serum sodium are more strongly correlated.
```

### Create a Bar Graph with Vertical Lines to Indicate Ranges

```python
plt.figure(figsize = (10, 6))
plt.axvline(50, color ='red')
plt.axvline(70, color ='red')

# The code above plt.axvline is what is indicating the red line, the 50 in the first one will place the line at the 50 on the x axis
# The 70 on the second line indicates where to put the other red line, on the 70 on the x axis
# color= 'red' is what is showing the line to be red, you can change it to blue, purple, etc, any color:)

sns.histplot(HeartFail, x= 'ejection_fraction', kde = True, hue='sex', color ='orange')

# The x= is showing the variable  of 'ejection_fraction', the hue= is set on the 'sex' variable to display the difference between
# genders.
# I wanted to plot these lines to show the normal range of ejection fraction is in between 50% and 70%. This shows how much of the
# dataset was outside the normal range.
# You can use these lines to indicate normal ranges of variables if your dataset has medical features or if your want to show
# changes across time.
```

### Violin Plot

```python
sns.violinplot(x=HeartFail.DEATH_EVENT, y=HeartFail.age, hue=HeartFail.sex, split=True)
```

```python
# The code above creates a violin plot! Aren't they so cute?
# The sns in the beginning is showing you it is from the seaborn package.The hue is showing the genders and the little key shows
# the blue is for females(0-I'm not a fan of females being coded as a zero but most data is like this) and the orange is males(1).
# The syntax is as follows if you want to use this to create your own violin plot!
# sns.violinplot(x=dataset.variable_column, y=dataset.variable_column, hue=dataset.variable_column, split=True)
# Do you see how this syntax is working in the code above? The x= is for the DEATH_EVENT the Dependent Variable
# The y= is looking at one of the independent variables of age, hue= is looking at the gender, another independent variable yet 
# notice how x= and y= are continuous variables while hue= is a categorical variable of gender meaning either male or female 
# categories. The split= True is how the gender is split in the violin. 
# Have fun trying this out on your own dataset!
```

### Create Beautiful Histograms 

```python
fig, axs = plt.subplots(3, 2, figsize=(15, 10))

sns.histplot(data=HeartFail, x="age", kde=True, color="skyblue", ax=axs[0, 0])
sns.histplot(data=HeartFail, x="creatinine_phosphokinase", kde=True, color="olive", ax=axs[0, 1])
sns.histplot(data=HeartFail, x="ejection_fraction", kde=True, color="gold", ax=axs[1, 0])
sns.histplot(data=HeartFail, x="platelets", kde=True, color="teal", ax=axs[1, 1])
sns.histplot(data=HeartFail, x="serum_sodium", kde=True, color="pink", ax=axs[2, 0])
sns.histplot(data=HeartFail, x="serum_creatinine", kde=True, color="blue", ax=axs[2, 1])
plt.show()
######excellent site where I learned some of this code
#https://www.python-graph-gallery.com/25-histogram-with-several-variables-seaborn
```
